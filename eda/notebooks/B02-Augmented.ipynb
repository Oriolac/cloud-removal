```python
import pandas as pd
import re
```

```python
augmented_df = pd.read_pickle("data/input/ts_pairings_1.pkl")
df = pd.read_pickle("data/input/cr/train.pkl")
augmented_df.shape[0], df.shape[0]
```

```python
rename_cols = {
    "cloudy_img_path": "cloudly_img_path"
}
```

```python
augmented_df = augmented_df.rename(columns=rename_cols)
```

```python
def add_path_root(type_, col):
    def add_type(x):
        m = re.match(f"s{type_}" + r"_(?P<roi>ROIs\d\d\d\d)_(?P<scene>\d+)_ImgNo_(?P<time_interval>\d+)_\d+-\d+-\d+_patch_(?P<patch>\d+)", x[col])
        try:
            _, _, ti, _ = m.groups()
        except AttributeError as e:
            print(x[col])
            raise e
        return x.continent + "/" + x.name[2] + f"/{x.name[0]}/S{type_}/{ti}/{x[col]}"
    return add_type
for s, col in zip([2,1,2], augmented_df.columns[:3]):
    augmented_df[col] = augmented_df.apply(add_path_root(s, col), axis=1)
augmented_df.head()
```

```python
res_df = pd.concat([augmented_df, df])
res_df
```

```python
res_df.to_pickle("data/input/augmented/train.pkl")
```
