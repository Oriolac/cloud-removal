```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-white")

df = pd.concat([pd.read_csv(r'C:\Users\oriol.alas\Downloads\metrics (6).csv'), pd.read_csv(r'C:\Users\oriol.alas\Downloads\metrics (5).csv')])

DICT = {
    "norm_test_loss": "MSE Regina",
    "norm_train_loss": "Train loss",
    "norm_val_loss": "Val loss",
    "accuracy_discriminator": "",
    "batch_fake": "Fake score",
    "batch_real": "Real score",
    "test_MSELoss": "MSE RegiGAN"
}

fig, ax = plt.subplots(figsize=(16, 4))
for name, gdf in df.groupby('key'):
    ax.plot(gdf.index, gdf['value'], label=DICT[name])
ax.legend()
ax.grid()
ax.set_xlabel("Batch")
ax.set_ylabel("Value")
ax.set_title("Scores of GAN when stabilized")
ax.set_xlim([0, 13])
fig.savefig(r"C:\Users\oriol.alas\PROJECTS\LACLASSE\docs\imgs\models\models\cnn\gan-mse.png")
```
